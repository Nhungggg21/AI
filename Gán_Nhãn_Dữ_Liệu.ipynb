```python
#Khai báo thư viện
import pandas as pd

# Bước 1: Đọc file CSV với xử lý lỗi mã hóa
df = pd.read_csv('/content/drive/MyDrive/Nhóm_3_BTL_Al/Data/path_to_save_file.csv', encoding='utf-8')

print("Dữ liệu gốc:")
print(df.head())

# Chuyển đổi cột Rating sang kiểu số (float)
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Từ khóa tích cực và tiêu cực
positive_keywords = ["ngon", "tuyệt vời", "hảo hạng", "đậm đà", "hấp dẫn", "thơm ngon", "tinh tế", "chất lượng", "đỉnh cao", "bổ dưỡng","perfect","hài lòng","hợp","quá tuyệt vời","chuẩn vị","giá rẻ","thơm","QUÁ NHIỀU","sạch sẽ","đầy đặn", "giá cả hợp lý", "good", "chu đáo", "tinh tế", "sạch sẽ", "thân thiện", "thái độ tốt"]
negative_keywords = ["dở", "tệ", "nhạt", "chán", "khô", "mặn", "nhão", "đắt", "ít", "không ngon","bực mình","chua","lởn cởn","không hợp","kém","chát","thiu","thất vọng", "phục vụ kém", "bẩn", "dơ", "khó chịu", "dai", "sống", "chưa chín", "không hợp vệ sinh", "mất vệ sinh","ghét, không thích, không vui, quá đắt, quá ít "]
# Định nghĩa hàm gán nhãn sentiment
def review_classification(review, rating):
    review_lower = str(review).lower()  # Chuyển tất cả các từ thành chữ thường
    review_words = review_lower.split()  # Tách các từ ra
    positive_count = sum([1 for word in review_words if word in positive_keywords])
    negative_count = sum([1 for word in review_words if word in negative_keywords])
    if positive_count > negative_count:
        return 'Positive'
    elif negative_count > positive_count:
        return 'Negative'
    else:
        if rating >= 7.6:
            return 'Positive'
        elif rating <= 5.5:
            return 'Negative'
        else:
            return 'Neutral'

# Gán nhãn sentiment cho các đánh giá
df['sentiment'] = df.apply(lambda row: review_classification(row['Comment'], row['Rating']), axis=1)

print("Dữ liệu sau khi gán nhãn:")
print(df.head(10))

# Bước 3: Lưu lại file CSV mới
df.to_csv('/content/drive/MyDrive/Nhóm_3_BTL_Al/Data/Labeldata_btl.csv', index=False)
print("Đã lưu file Labeldata_btl.csv")
```
